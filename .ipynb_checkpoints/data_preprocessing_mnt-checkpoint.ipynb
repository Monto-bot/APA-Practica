```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.impute import SimpleImputer
import missingno as msno
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
```

```python
exoplanets_raw = pd.read_csv('exoplanets.csv')
print(f"Dimensiones: {exoplanets_raw.shape}")
print(exoplanets_raw.head().transpose())

metadata = pd.read_csv('columns.csv')
assert set(exoplanets_raw.columns) == set(metadata['column_name']) # verfificamos que metadata esté completa
targets_mask = metadata['Target'] == 'Yes'
features_mask = metadata['Feature'] == 'Yes'
```

```python
features = exoplanets_raw.columns[features_mask]
targets = exoplanets_raw.columns[targets_mask]
exoplanets = exoplanets_raw[features.union(targets)]

X = exoplanets[features]
Y = exoplanets[targets]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42, stratify = Y)
print(f"Train: {X_train.shape}, Test: {X_test.shape}")
```

## Paso de incertidumbres absolutas a relativas


Para no tener problemas con divisiones por cero, que solo ocurren en la columna "koi_impact", le sumamos una constante antes de hacer la conversión a relativa. Así, la dispersion absoluta no cambia, pero evitamos divisiones por cero. La dispersion relativa se ve afectada, pero sigue codificando la misma información.

```python
X_train['koi_impact'] += 1
X_test['koi_impact'] += 1

for col in X_train.columns:
    if "_err" in col:
        origin_col = col.split("_err")[0]
        if origin_col in X_train.columns:
            X_train[col] = X_train[col]/X_train[origin_col]
            X_test[col] = X_test[col]/X_test[origin_col]
```

## Imputación de valores faltantes

```python
print(f"--- Valores faltantes ---")
print(f"Train: {X_train.isnull().sum().sum()} valores nulos")
print(f"Test: {X_test.isnull().sum().sum()} valores nulos")
```

```python
msno.matrix(X_train);
plt.title("Missing values - Train")
plt.xlabel("Variables (columnas)", fontsize = 12)
plt.ylabel("Observaciones (filas)", fontsize = 12)
plt.show()

msno.matrix(X_test);
plt.title("Missing values - Test")
plt.xlabel("Variables (columnas)", fontsize = 12)
plt.ylabel("Observaciones (filas)", fontsize = 12)
plt.show()

null_map_train = X_train.isnull().sum(axis = 1)
plt.figure()
plt.hist(null_map_train, bins = 30)
plt.title("Distribución de nulos por fila - Train")
plt.xlabel("Número de valores nulos por fila", fontsize = 12)
plt.ylabel("Número de filas (frecuencia)", fontsize = 12)
plt.show()

null_map_test = X_train.isnull().sum(axis = 1)
plt.figure()
plt.hist(null_map_test, bins = 30)
plt.title("Distribución de nulos por fila - Test")
plt.xlabel("Número de valores nulos por fila", fontsize = 12)
plt.ylabel("Número de filas (frecuencia)", fontsize = 12)
plt.show()
```

```python
rows = null_map_train[null_map_train > 10].index

print("="*70)
print("Distribución de variable objetivo en filas con >10 valores faltantes:")
print("="*70)
print(exoplanets.iloc[rows, :].loc[:,"koi_disposition"].value_counts())
print(f"\nTotal de filas con >10 nulos: {len(rows)}")

print("\n" + "="*70)
print("Distribución de variable objetivo en todo el dataset:")
print("="*70)
print(exoplanets.loc[:,"koi_disposition"].value_counts())
print(f"\nTotal de filas en el dataset: {len(exoplanets)}")
```

```python
finfo = np.finfo(np.float64)
mask_too_big = (X_train > finfo.max) | (X_train < finfo.min)

print("="*70)
print("Columnas con almenos un valor fuera del rango representable de float64")
print("="*70)
mask_too_big.any().sum()
```

```python
X_train.drop(columns = ['koi_teq_err1', 'koi_teq_err2'], inplace = True)
X_test.drop(columns = ['koi_teq_err1', 'koi_teq_err2'], inplace = True)
print(f"Eliminadas columnas: {['koi_teq_err1', 'koi_teq_err2']}")
```

```python
X_train["null_count_flag"] = (null_map_train > 10).astype(int)
X_test["null_count_flag"] = (null_map_train > 10).astype(int)

print(f"Filas con >10 nulos - Train: {X_train['null_count_flag'].sum()}")
print(f"Filas con >10 nulos - Test: {X_test['null_count_flag'].sum()}")
```

```python
imputer = SimpleImputer(strategy='median')

X_train = pd.DataFrame(imputer.fit_transform(X_train), columns = X_train.columns, index = X_train.index)
X_test = pd.DataFrame(imputer.transform(X_test), columns = X_test.columns, index = X_test.index)

# verificación que no hay missing values
assert X_train.isnull().sum().sum() == 0
assert X_test.isnull().sum().sum() == 0
# visualmente
msno.matrix(X_train);
msno.matrix(X_test);
```

## Tratamiento de outliers

Primero, pasamos todos los datos a positivos para evitar problemas con el clipping superior

```python
negative_counts = (X_train < 0).sum()
cols_with_negatives = negative_counts[negative_counts > 0].index

print("Columnas con valores negativos:")
non_pos = (X_train[cols_with_negatives] <= 0).sum()
print(non_pos)
```

```python
X_train = abs(X_train)
X_test = abs(X_test)
```

Como no podemos garantizar que los valores atípicos detectados sean errores de medición, no los eliminamos del dataset. Para no romper la escala, los tratamos mediante winsorización.

```python
fig, axes = plt.subplots(len(X_train.columns), 2, figsize = (10, len(X_train.columns)*2))
for i, col in enumerate(X_train.columns):
    sns.histplot(x = X_train[col], ax = axes[i,0])
    sns.scatterplot(x = X_train[col], y = Y_train['koi_disposition'], ax = axes[i,1])
plt.tight_layout()
plt.show()
```

```python
extreme_cols = ["koi_period", "koi_period_err1", "koi_period_err2",
                "koi_prad", "koi_prad_err1", "koi_prad_err2",
                "koi_insol", "koi_insol_err1",
                "koi_srad", "koi_srad_err1", "koi_srad_err2",
                "koi_depth", "koi_depth_err1", "koi_depth_err2"]
skewed_cols = ["koi_impact_err2", "koi_impact_err1", "koi_impact",
               "koi_duration", "koi_duration_err1", "koi_duration_err2",
               "koi_teq",
               "koi_model_snr"]

upper_bounds = X_train[extreme_cols].quantile(0.99)
X_train[extreme_cols] = X_train[extreme_cols].clip(upper = upper_bounds, axis = 1)
X_test[extreme_cols] = X_test[extreme_cols].clip(upper = upper_bounds, axis = 1)
```

```python
lof = LocalOutlierFactor(n_neighbors=20)
result = lof.fit_predict(X_train)
outliers = result == -1
print(len(outliers))
```

Veiem que Local Outlier Factor no ens funciona bé degut a la gran quantitat de dades que tenim i la seva alta dimensionalitat.

```python
fig, axes = plt.subplots(len(X_train.columns), 2, figsize=(10, len(X_train.columns)*2))
for i, col in enumerate(X_train.columns):
    sns.histplot(x=X_train[col], ax=axes[i,0])
    sns.scatterplot(x=X_train[col], y=Y_train['koi_disposition'], ax=axes[i,1])
plt.tight_layout()
plt.show()
```

```python
X_train[extreme_cols + skewed_cols] = np.log1p(X_train[extreme_cols + skewed_cols])
X_test[extreme_cols + skewed_cols] = np.log1p(X_test[extreme_cols + skewed_cols])
```

```python
fig, axes = plt.subplots(len(X_train.columns), 2, figsize=(10, len(X_train.columns)*2))
for i, col in enumerate(X_train.columns):
    sns.histplot(x=X_train[col], ax=axes[i,0])
    sns.scatterplot(x=X_train[col], y=Y_train['koi_disposition'], ax=axes[i,1])
plt.tight_layout()
plt.show()
```

## Análisis de correlaciones

```python
print("\n--- Análisis de correlaciones ---")
corr_mat = X_train.corr()

plt.figure(figsize = (16, 14))
sns.heatmap(corr_mat, cmap = 'coolwarm', center = 0, square = True, linewidths = 0.5, cbar_kws = {"shrink": 0.8})
plt.title("Matriz de Correlación - Features")
plt.tight_layout()
plt.show()

# Identificar correlaciones altas
high_corr = np.where(np.abs(corr_mat) > 0.9)
high_corr_list = [(corr_mat.index[x], corr_mat.columns[y], corr_mat.iloc[x, y]) for x, y in zip(*high_corr) if x != y and x < y]

print(f"\nPares de variables con correlación > 0.9:")
for var1, var2, corr in high_corr_list[:10]:
    print(f"  {var1} <-> {var2}: {corr:.3f}")
```

## Seleccion de variables amb PCA

```python
sscaler = StandardScaler()
X_train_ss = sscaler.fit_transform(X_train)
X_test_ss = sscaler.transform(X_test)

myPCA = PCA()
X_train_pca = myPCA.fit_transform(X_train_ss)
fig, ax = plt.subplots(1, 2, figsize = (12, 5))
sns.lineplot(x = range(len(myPCA.explained_variance_)), y = myPCA.explained_variance_ratio_.cumsum(), ax = ax[0])
sns.scatterplot(x = X_train_pca[:, 0], y = X_train_pca[:, 1], hue = Y_train['koi_disposition'], ax = ax[1])
```

```python
sns.heatmap(pd.DataFrame(data = myPCA.components_, columns = X_train.columns), cmap = 'viridis')
len(X_train.columns)
```

Como vemos, no hay relacion clara entre las dos primeras componentes principales y la variable objetivo. Además, aunque vemos que la varianza total puede ser explicada por unas 20 componentesles, no creemos que perder la interpretabilidad de las variables originales merezca la pena. Por tanto, decidimos no aplicar PCA para la reducción de dimensionalidad.


## Normalización de los datos

Como podemos observar en las gráficas anteriores, las variables no son gausianas y por tanto, la normalización mediante StandardScaler no es la más adecuada. En su lugar, optamos por usar MinMaxScaler para escalar las variables al rango [0, 1].

```python
from sklearn.preprocessing import MinMaxScaler
mmscaler = MinMaxScaler()
X_train_mm = pd.DataFrame(mmscaler.fit_transform(X_train), columns=X_train.columns)
X_test_mm = pd.DataFrame(mmscaler.transform(X_test), columns=X_test.columns)
```

## Guardado de los datos preprocesados

```python
X_train_mm.to_csv('X_train_preprocessed.csv', index=False)
X_test_mm.to_csv('X_test_preprocessed.csv', index=False)
Y_train.to_csv('Y_train.csv', index=False)
Y_test.to_csv('Y_test.csv', index=False)

print("✓ Datos preprocesados guardados")
print(f"  - X_train: {X_train_mm.shape}")
print(f"  - X_test: {X_test_mm.shape}")
print(f"  - Y_train: {Y_train.shape}")
print(f"  - Y_test: {Y_test.shape}")
```
