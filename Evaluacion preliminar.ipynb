## 1. Descripción del conjunto de datos y su fuente
Los datos fueron medidos por la misión Kepler (telescopio espacial de la NASA).

El objetivo es entrenar un modelo capaz de predecir si una observación es un candidato real a ser un exoplaneta o no.
OPCIO A: Por tanto, este es un problema de classificación, donde la variable objetivo es decidir a que categoria perteneze cada entrada: Candidato o No Candidato.
OPCIO B: Por tanto, este es un problema de regressión, donde la variable objetivo es el porcentage de certidumbre de sí el objeto es Candidato o No Candidato.

- **kepid:** Número de identificación Kepler (KepID). Identificador único del objetivo en el Kepler Input Catalog (KIC). Es un número entero de 9 dígitos.
- **kepoi_name:** Nombre del Objeto de Interés Kepler (KOI). Identificador en formato KIC (con un '.01', '.02', ... al final para sistemas con múltiples candidatos).
- **kepler_name:** Nombre del planeta confirmado por Kepler. Asignado por el equipo de Kepler a planetas confirmados/validados. Formato: 'Kepler-' + número + letra (ej: Kepler-11 b). Para candidatos no confirmados, el valor es 'NaN'.
- **koi_disposition:** Disposición del KOI en el Exoplanet Archive. Categoría asignada basada en la evidencia disponible: CANDIDATE (señal consistente con planeta), FALSE POSITIVE (señal explicada por otros fenómenos), NOT DISPOSITIONED (análisis incompleto), o CONFIRMED (confirmado por métodos independientes).
- **koi_pdisposition:** Disposición del KOI usando solo datos de Kepler. Representa la explicación física más probable antes de considerar observaciones de seguimiento. Puede diferir de koi_disposition cuando hay datos externos.
- **koi_score:** Puntuación de disposición del KOI. Valor entre 0 y 1 que indica la confianza en que la señal sea planetaria, donde 1 es máxima confianza. Basado en un clasificador automático que evalúa múltiples características del tránsito.
- **koi_fpflag_nt:** Indica si la curva de luz es inconsistente con un tránsito planetario. Valores: 1 (problemas de forma/fase), 0 (forma de tránsito aceptable)."
- **koi_fpflag_ss:** Señala si el KOI probablemente corresponde a un sistema binario eclipsante.
- **koi_fpflag_co:** Indica si la fuente de la señal está en una estrella cercana diferente al objetivo, basado en análisis de desplazamiento del centroide durante el tránsito.
- **koi_fpflag_ec:** Detecta contaminación por coincidencia con tránsitos de otros KOIs conocidos en el mismo campo, sugiriendo contaminación cruzada.
- **koi_period:** Periodo orbital en días. Tiempo entre tránsitos consecutivos medido por ajuste de modelo de tránsito a los datos de fotometría de Kepler.
- **koi_period_err1:** Incertidumbre positiva (+1σ) en el periodo orbital, en días. Representa el percentil 84 de la distribución posterior del parámetro.
- **koi_period_err2:** Incertidumbre negativa (-1σ) en el periodo orbital, en días. Representa el percentil 16 de la distribución posterior del parámetro.
- **koi_time0bk:** Época de tránsito de referencia. Tiempo del centro del primer tránsito detectado, expresado en BKJD (Besselian Kepler Julian Date).
- **koi_time0bk_err1:** Incertidumbre positiva (+1σ) en la época de tránsito, en días BKJD.
- **koi_time0bk_err2:** Incertidumbre negativa (-1σ) en la época de tránsito, en días BKJD.
- **koi_impact:** Parámetro de impacto. Distancia normalizada entre la proyección del planeta sobre el disco estelar y el centro de la estrella, en unidades de radios estelares. Rango: 0 (tránsito central) a 1 (tránsito tangencial).
- **koi_impact_err1:** Incertidumbre positiva (+1σ) en el parámetro de impacto.
- **koi_impact_err2:** Incertidumbre negativa (-1σ) en el parámetro de impacto.
- **koi_duration:** Duración del tránsito en horas. Intervalo de tiempo desde el primer hasta el último contacto del tránsito, incluyendo los efectos de limb-darkening.
- **koi_duration_err1:** Incertidumbre positiva (+1σ) en la duración del tránsito, en horas.
- **koi_duration_err2:** Incertidumbre negativa (-1σ) en la duración del tránsito, en horas.
- **koi_depth:** Profundidad del tránsito en partes por millón (ppm). Diferencia fraccional en flujo entre fuera del tránsito y el mínimo del tránsito: (F_out - F_in)/F_out × 10^6.
- **koi_depth_err1:** Incertidumbre positiva (+1σ) en la profundidad del tránsito, en ppm.
- **koi_depth_err2:** Incertidumbre negativa (-1σ) en la profundidad del tránsito, en ppm.
- **koi_prad:** Radio planetario en radios terrestres. Radio físico del planeta derivado de la profundidad del tránsito y parámetros estelares. Rt = 6,371 km.
- **koi_prad_err1:** Incertidumbre positiva (+1σ) en el radio planetario, en radios terrestres.
- **koi_prad_err2:** Incertidumbre negativa (-1σ) en el radio planetario, en radios terrestres.
- **koi_teq:** Temperatura de equilibrio del planeta en Kelvin. Temperatura que tendría el planeta asumiendo albedo Bond cero y redistribución de calor completa: T_eq = T_eff × √(R_star/2a).
- **koi_teq_err1:** Incertidumbre positiva (+1σ) en la temperatura de equilibrio, en Kelvin.
- **koi_teq_err2:** Incertidumbre negativa (-1σ) en la temperatura de equilibrio, en Kelvin.
- **koi_insol:** Insolación planetaria en unidades terrestres. Flujo de radiación recibido por el planeta dividido por el flujo solar recibido por la Tierra en la parte superior de la atmósfera (1361 W/m²).
- **koi_insol_err1:** Incertidumbre positiva (+1σ) en la insolación planetaria, en unidades terrestres.
- **koi_insol_err2:** Incertidumbre negativa (-1σ) en la insolación planetaria, en unidades terrestres.
- **koi_model_snr:** Relación señal-ruido del modelo de tránsito. SNR de la señal de tránsito calculada como la profundidad dividida por la incertidumbre en la profundidad, derivada del ajuste del modelo.
- **koi_count:** Número de candidatos planetarios en el sistema. Cuenta total de KOIs asociados al mismo Kepid (sistema estelar), indicando la multiplicidad del sistema.
- **koi_tce_plnt_num:** Número de planeta en el TCE. Identificador secuencial del planeta dentro del conjunto de Threshold Crossing Events federado al KOI.
- **koi_tce_delivname:** Nombre de la entrega TCE. Identifica la versión específica del procesamiento de datos que produjo el Threshold Crossing Event federado al KOI."
- **koi_steff:** Temperatura efectiva estelar en Kelvin. Temperatura de un cuerpo negro que emitiría la misma radiación total por unidad de área superficial que la estrella.
- **koi_steff_err1:** Incertidumbre positiva (+1σ) en la temperatura efectiva estelar, en Kelvin.
- **koi_steff_err2:** Incertidumbre negativa (-1σ) en la temperatura efectiva estelar, en Kelvin.
- **koi_slogg:** Gravedad superficial estelar en log10(cm/s²). Logaritmo base 10 de la aceleración gravitacional en la superficie de la estrella.
- **koi_slogg_err1:** Incertidumbre positiva (+1σ) en la gravedad superficial estelar, en log10(cm/s²).
- **koi_slogg_err2:** Incertidumbre negativa (-1σ) en la gravedad superficial estelar, en log10(cm/s²).
- **koi_srad:** Radio estelar en radios solares. Radio físico de la estrella derivado de modelos evolutivos estelares o relaciones escala. Rs = 696,342 km.
- **koi_srad_err1:** Incertidumbre positiva (+1σ) en el radio estelar, en radios solares.
- **koi_srad_err2:** Incertidumbre negativa (-1σ) en el radio estelar, en radios solares.
- **ra:** Ascensión recta en grados. Coordenada ecuatorial celeste, análoga a la longitud terrestre. Rango: 0° a 360°.
- **dec:** Declinación en grados. Coordenada ecuatorial celeste, análoga a la latitud terrestre. Rango: -90° a +90°.
- **koi_kepmag:** Magnitud Kepler. Magnitud aparente de la estrella en la banda de paso del instrumento Kepler (aproximadamente 430-890 nm). Sistema fotométrico AB.

## 3. Matriz de correlación de las variables numéricas

```python
# Seleccionar solo columnas numéricas
numeric_columns = exo.select_dtypes(include=['number']).columns
exo_numeric = exo[numeric_columns]

# Calcular correlación
corr = exo_numeric.corr()

# Mostrar resultados
display(corr)
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Matriz de Correlación (Solo Variables Numéricas)')
plt.tight_layout()
plt.show()
```
